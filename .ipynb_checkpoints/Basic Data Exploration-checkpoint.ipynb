## TIME TO PROVE MY FIRST ML MODEL -- HOME DATA FROM MELBOURNE

```python
import pandas as pd
```

```python
melbourne_data = pd.read_csv("melb_data.csv")
```

```python
melbourne_data.describe()
```

```python
melbourne_data.head()
```

```python
melbourne_data.columns
```

```python
melbourne_data = melbourne_data.dropna(axis = 0)
```

```python
melbourne_data
```

### Start to select the prediction target --> we only want the Price in the
### Melbourne data

```python
y = melbourne_data.Price
```

### Choosing "Feauteres" 

```python
melbourne_feature = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
```

```python
X = melbourne_data[melbourne_feature]
```

```python
X.describe()
```

```python
X.head()
```

## Bulding our model --> use scikit-learn to create my models. 

### Steps:
- Define : in  this case is a decision tree
- Fit: Capture patterns from provided data (heart of modeling)
- Predict
- Evaluate: Determine how accurate the model's prediction are

```python
from sklearn.tree import DecisionTreeRegressor

melbourne_model = DecisionTreeRegressor(random_state = 1)

# Fit the model 

melbourne_model.fit(X, y);
```

```python
print("Making predictions for the following 5 houses: ")
print(X.head())

print("The predictions are")
print(melbourne_model.predict(X.head()))
```
